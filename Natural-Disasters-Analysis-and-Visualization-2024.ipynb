```python
# Gerekli kütüphaneler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Grafikler için görsel stili ayarladık
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Veri setini yükledik
data = pd.read_csv('/natural_disasters_2024.csv')

# Veri seti hakkında temel bilgileri yazdırdık
print("Veri Seti Genel Bakış:")
data.info()
print("\nİlk 5 satır:\n", data.head())

# Analiz 1: En sık görülen afet türlerini yazdırdık
en_sik_afet_turleri = data['Disaster_Type'].value_counts()
print("\nEn Sık Görülen Afet Türleri:\n", en_sik_afet_turleri)

# Görselleştirme 1: En sık görülen afet türleri için pasta grafiği oluşturduk
en_sik_afet_turleri.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title("En Sık Görülen Afet Türleri")
plt.ylabel("")
plt.show()

# Analiz 2: Afet türlerine göre ekonomik kayıpları yazdırdık
ekonomik_kayip_afet_turu = data.groupby('Disaster_Type')['Economic_Loss($)'].sum().sort_values(ascending=False)
print("\nAfet Türlerine Göre Ekonomik Kayıp:\n", ekonomik_kayip_afet_turu)

# Görselleştirme 2: Afet türlerine göre ekonomik kayıp için çubuk grafiği oluşturduk
ekonomik_kayip_afet_turu.plot.bar(color=sns.color_palette("muted"))
plt.title("Afet Türlerine Göre Ekonomik Kayıp")
plt.xlabel("Afet Türü")
plt.ylabel("Ekonomik Kayıp ($)")
plt.show()

# Analiz 3: Afet türlerine göre ölüm sayılarını yazdırdık
oelum_afet_turu = data.groupby('Disaster_Type')['Fatalities'].sum().sort_values(ascending=False)
print("\nAfet Türlerine Göre Ölüm Sayıları:\n", oelum_afet_turu)

# Görselleştirme 3: Afet türlerine göre ölüm sayıları için çubuk grafiği oluşturduk
oelum_afet_turu.plot.bar(color=sns.color_palette("dark"))
plt.title("Afet Türlerine Göre Ölüm Sayıları")
plt.xlabel("Afet Türü")
plt.ylabel("Toplam Ölüm")
plt.show()

# Analiz 4: Zaman içerisindeki afet sayısını yazdırdık
data['Year'] = pd.to_datetime(data['Date']).dt.year
yillik_afet_sayisi = data['Year'].value_counts().sort_index()
print("\nYıllık Afet Sayısı:\n", yillik_afet_sayisi)

# Görselleştirme 4: Yıllık afet sayısı için çizgi grafiği oluşturduk
yillik_afet_sayisi.plot(marker='o', color="skyblue")
plt.title("Yıllık Afet Sayısı")
plt.xlabel("Yıl")
plt.ylabel("Afet Sayısı")
plt.grid(True)
plt.show()

# Analiz 5: Konum bazında ortalama büyüklüklerini yazdırdık
ortalama_buyukluk_konum = data.groupby('Location')['Magnitude'].mean().sort_values(ascending=False)
print("\nKonum Bazında Ortalama Büyüklük:\n", ortalama_buyukluk_konum)

# Görselleştirme 5: Konum bazında ortalama büyüklük için çubuk grafiği oluşturduk
ortalama_buyukluk_konum.plot.bar(color=sns.color_palette("husl"))
plt.title("Konum Bazında Ortalama Büyüklük")
plt.xlabel("Konum")
plt.ylabel("Ortalama Büyüklük")
plt.xticks(rotation=45)
plt.show()

# Analiz 6: Büyüklük ile ekonomik kayıp arasındaki korelasyonu yazdırdık
korelasyon_buyukluk_ekonomik_kayip = data[['Magnitude', 'Economic_Loss($)']].corr().iloc[0, 1]
print("\nBüyüklük ile Ekonomik Kayıp Arasındaki Korelasyon:\n", korelasyon_buyukluk_ekonomik_kayip)

# Görselleştirme 6: Büyüklük ile ekonomik kayıp için dağılım grafiği oluşturduk
sns.scatterplot(data=data, x="Magnitude", y="Economic_Loss($)", alpha=0.6, color="coral")
plt.title("Büyüklük ve Ekonomik Kayıp")
plt.xlabel("Büyüklük")
plt.ylabel("Ekonomik Kayıp ($)")
plt.show()

# Analiz 7: Her afet türünün toplam ortalama büyüklüklerini yazdırdık
ortalama_buyukluk_afet_turu = data.groupby('Disaster_Type')['Magnitude'].mean().sort_values(ascending=False)
print("\nAfet Türlerine Göre Ortalama Büyüklük:\n", ortalama_buyukluk_afet_turu)

# Görselleştirme 7: Afet türlerine göre ortalama büyüklükleri çubuk grafiği ile oluşturduk
ortalama_buyukluk_afet_turu.plot.bar(color=sns.color_palette("cool"))
plt.title("Afet Türlerine Göre Ortalama Büyüklük")
plt.xlabel("Afet Türü")
plt.ylabel("Ortalama Büyüklük")
plt.xticks(rotation=45)
plt.show()

# Analiz 8: Yıllara göre afet türlerinin toplam sayısını yazdırdık
yillik_afet_turu_sayisi = data.groupby(['Year', 'Disaster_Type']).size().unstack(fill_value=0)
print("\nYıllara Göre Afet Türlerinin Toplam Sayısı:\n", yillik_afet_turu_sayisi)

# Görselleştirme 8: Yıllara göre afet türlerinin toplam sayısı için yığılmış çubuk grafiği  oluşturduk
yillik_afet_turu_sayisi.plot(kind='bar', stacked=True, colormap='viridis')
plt.title("Yıllara Göre Afet Türlerinin Toplam Sayısı")
plt.xlabel("Yıl")
plt.ylabel("Afet Sayısı")
plt.legend(title="Afet Türü")
plt.show()

# Analiz 9: En yüksek ekonomik kaybın yaşandığı ilk 10 afeti yazdırdık
en_yuksek_kayip_afetler = data.nlargest(10, 'Economic_Loss($)')
print("\nEn Yüksek Ekonomik Kayıp Yaşanan 10 Afet:\n", en_yuksek_kayip_afetler)

# Görselleştirme 9: En yüksek ekonomik kayıp yaşanan afetler için çubuk grafiği oluşturduk
en_yuksek_kayip_afetler.set_index('Disaster_ID')['Economic_Loss($)'].plot(kind='bar', color="salmon")
plt.title("En Yüksek Ekonomik Kayıp Yaşanan 10 Afet")
plt.xlabel("Afet ID")
plt.ylabel("Ekonomik Kayıp ($)")
plt.show()

# Analiz 10: Ekonomik kayıp ve ölüm sayısı arasındaki ilişkileri yazdırdık
korelasyon_kayip_olum = data[['Economic_Loss($)', 'Fatalities']].corr().iloc[0, 1]
print("\nEkonomik Kayıp ve Ölüm Sayısı Arasındaki Korelasyon:\n", korelasyon_kayip_olum)

# Görselleştirme 10: Ekonomik kayıp ve ölüm sayısı için dağılım grafiği oluşturduk
sns.scatterplot(data=data, x="Economic_Loss($)", y="Fatalities", alpha=0.6, color="darkblue")
plt.title("Ekonomik Kayıp ve Ölüm Sayısı")
plt.xlabel("Ekonomik Kayıp ($)")
plt.ylabel("Ölüm Sayısı")
plt.show()

# Analiz 11: En fazla ölüm oranı olan ilk 10 afeti yazdırdık

data['Fatality_Rate'] = data['Fatalities'] / data['Magnitude']
en_fazla_olum_orani_afetler = data.nlargest(10, 'Fatality_Rate')
print("\nEn Fazla Ölüm Oranı Olan 10 Afet:\n", en_fazla_olum_orani_afetler)



# Analiz 12: Afet türlerine göre ölüm oranlarını yazdırdık
ortalama_olum_oranlari = data.groupby('Disaster_Type')['Fatality_Rate'].mean().sort_values(ascending=False)
print("\nAfet Türlerine Göre Ortalama Ölüm Oranı:\n", ortalama_olum_oranlari)



# Analiz 13: Konum bazında toplam ölüm sayılarını yazdırdık
konum_bazinda_olum = data.groupby('Location')['Fatalities'].sum().sort_values(ascending=False)
print("\nKonum Bazında Toplam Ölüm Sayıları:\n", konum_bazinda_olum)



# Analiz 14: Ekonomik kayıpların yıllık toplamını yazdırdık
yillik_ekonomik_kayip = data.groupby(data['Date'].str[:4])['Economic_Loss($)'].sum()
print("\nYıllık Ekonomik Kayıp Toplamı:\n", yillik_ekonomik_kayip)



# Analiz 15: Her afet türünün medyan büyüklüklerini yazdırdık
medyan_buyukluk_afet_turu = data.groupby('Disaster_Type')['Magnitude'].median().sort_values(ascending=False)
print("\nAfet Türlerine Göre Medyan Büyüklük:\n", medyan_buyukluk_afet_turu)



# Analiz 16: En sık afetlerin meydana geldiği ilk 10 konumu yazdırdık
en_sik_afet_konumlari = data['Location'].value_counts().head(10)
print("\nEn Sık Afetlerin Meydana Geldiği İlk 10 Konum:\n", en_sik_afet_konumlari)



# Analiz 17: Büyüklük aralıklarına göre afet dağılımını yazdırdık
buyukluk_kategorileri = pd.cut(data['Magnitude'], bins=[0, 3, 6, 9, 12], labels=['Düşük', 'Orta', 'Yüksek', 'Çok Yüksek'])
buyukluk_kategorileri_sayisi = buyukluk_kategorileri.value_counts()
print("\nBüyüklük Aralıklarına Göre Afet Dağılımı:\n", buyukluk_kategorileri_sayisi)



# Analiz 18: En fazla büyüklük gösteren ilk 10 afeti yazdırdık
en_fazla_buyukluk_afetler = data.nlargest(10, 'Magnitude')
print("\nEn Fazla Büyüklük Gösteren İlk 10 Afet:\n", en_fazla_buyukluk_afetler)



# Analiz 19: Konum bazında toplam ekonomik kayıbı yazdırdık
konum_bazinda_kayip = data.groupby('Location')['Economic_Loss($)'].sum().sort_values(ascending=False).head(10)
print("\nKonum Bazında Toplam Ekonomik Kayıp:\n", konum_bazinda_kayip)



# Analiz 20: Yıllara göre büyüklük ortalamalarını yazdırdık
yillik_buyukluk_ortalama = data.groupby(data['Date'].str[:4])['Magnitude'].mean()
print("\nYıllara Göre Büyüklük Ortalamaları:\n", yillik_buyukluk_ortalama)



```
